```python
import ipywidgets as widgets
from IPython.display import display
import time
from transformers import T5ForConditionalGeneration, AutoTokenizer

class TranslationInterface:
    def __init__(self):
        self.heading = widgets.HTML(value="<h2>GPT powered multilingual translation system</h2>")

        self.input_text = widgets.Textarea(description="Input Text to be translated:", layout=widgets.Layout(width='400px'))

        self.input_language = widgets.Dropdown(options=["English", "French", "German"], description="Input text language:")

        self.output_language = widgets.Dropdown(options=["English", "French", "German"], description="Output text language:")

        self.tokenization = widgets.Dropdown(options=["GPT", "Google Flan", "BERT"], description="Tokenization:")

        self.model = widgets.Dropdown(options=["Google Flan", "GPT", "BERT"], description="Model:")

        self.running_progress = widgets.IntProgress(min=0, max=100, description="Running:")

        self.submit_button = widgets.Button(description="Submit")
        self.submit_button.on_click(self.submit_translation)

        self.output_text = widgets.Textarea(description="Output:")

        self.interface = widgets.VBox([self.heading, self.input_text, self.input_language, self.output_language,
                                        self.tokenization, self.model, self.running_progress, self.submit_button, self.output_text])

        display(self.interface)

        self.model_name = "google/flan-t5-large"
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        self.model = T5ForConditionalGeneration.from_pretrained(self.model_name)

    def translate(self, input_text, src_lang, to_lang):
        prompt = f"Translate {src_lang} to {to_lang}: {input_text}"
        input_ids = self.tokenizer(prompt, return_tensors="pt").input_ids
        outputs = self.model.generate(input_ids, max_new_tokens=1000)
        model_translation = self.tokenizer.decode(outputs[0])
        final_translation = model_translation[5:-4]
        return final_translation

    def submit_translation(self, b):
        input_text = self.input_text.value
        input_language = self.input_language.value
        output_language = self.output_language.value
        tokenizer = self.tokenization.value

        # Simulate translation processing
        for i in range(101):
            time.sleep(0.1)
            self.running_progress.value = i

        # Call the translation function
        translated_text = self.translate(input_text, input_language, output_language)
        self.output_text.value = translated_text


if __name__ == "__main__":
    translator = TranslationInterface()
```
